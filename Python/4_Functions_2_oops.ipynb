```python
num=int(input("Enter the number:"))
def oddeven():
  if((num%2)==1):
    print("Odd number")
    message="Odd Number"
  else:
    print("Even number")
    message="Even Number"
  return message
oddeven()
```

```python
message=oddeven()
if(message=="Even Number"):
  print("The given number is even")
else:
  print("The given number is odd")
```

```python
def BMI():
  BMI_index=float(input("Enter the BMI Index:"))
  if(BMI_index<18.5):
    print("Under weight")
    message="Under weight"
  elif(BMI_index<24.9):
    print("Healthy weight")
    message="Healthy weight"
  elif(BMI_index<35):
    print("Very Overweight")
    message="Very Overweight"
  else:
    print("Higher weight status")
    message="Higher weight status"
  return message
BMI()
```
