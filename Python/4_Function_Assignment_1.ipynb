```python
### Question 3 ###

class EligibilityForMarriage():

    def Eligible():
        gender = input("Your Gender: ")
        age = int(input("Your Age: "))

        if((gender == "Male" and age >= 21) or
           (gender == "Female" and age >= 18)):
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")

EligibilityForMarriage.Eligible()
```
