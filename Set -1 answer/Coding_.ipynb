```python
'''5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number'''

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

num = 7
if is_prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")

```
