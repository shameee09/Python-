```python
## solution ##

age=int(input("Enter the age:"))
if(age<18):
  print("Children")
elif(age<35):
  print("Adult")
elif(age<59):
  print("Citizen")
else:
  print("Senior citizen")
```

```python
## for loop ##
lists =[23,34,59,78,18]
for temp in lists:
  print(temp)
```

```python
for age in lists:
   if(age<18):
       print("Children")
   elif(age<35):
       print("Adult")
   elif(age<59):
      print("Citizen")
   else:
      print("Senior citizen")
```
